##4. View Column Names

##6. Filter African Countries
This filters the dataset to show only African Countries and shows the dataset size and number of countries.

```python
african_countries=['Algeria','Angola','Benin','Botswana','Burkina Faso','Burundi','Cameroon','Cape Verde','Central African Republic','Chad','Comoros','Congo','Democratic Republic of Congo','Djibouti','Egypt','Equatorial Guinea','Eritrea','Eswatini','Ethiopia','Gabon','Gambia','Ghana','Guinea','Guinea-Bissau','Kenya','Lesotho','Liberia','Libya','Madagascar','Malawi','Mali','Mauritania','Mauritius','Morocco','Mozambique','Namibia','Niger','Nigeria','Rwanda','Senegal','Sierra Leone','Somalia','South Africa','South Sudan','Sudan','Tanzania','Togo','Tunisia','Uganda','Zambia','Zimbabwe']
africa_df=df[df['country'].isin(african_countries)]
print("African dataset shape:",africa_df.shape)
print("Number of countries:",africa_df['country'].nunique())
```

##7. Check Missing Values (Africa Data)
This checks for missing values in the African Dataset

```python
print(africa_df.isnull().sum())
```

##8. Select Relevant Column
Keep only the columns we need for analysis and preview the dataset

```python
cols=['country','date','new_cases','total_cases','new_deaths','total_deaths','new_cases_per_million','new_deaths_per_million','cfr']
africa_df=africa_df[cols]
print(africa_df.shape)
print(africa_df.head())
```

##9. Check CFR Column
Count how many CFR (Case Fatality Rate) values are missing and how many are available

```python
total_rows=len(africa_df)
cfr_nan=africa_df['cfr'].isnull().sum()
cfr_available=total_rows-cfr_nan
print(f"Total rows:{total_rows}")
print(f"CFR NaN values:{cfr_nan}")
print(f"CFR available values:{cfr_available}")
print(f"CFR NaN Percentage:{round((cfr_nan/total_rows)*100,2)}%")
```

##10. Handle Missing Value
Fill missing values in cases and death columns with zero.
Calculate missing CFR values from total deaths/total cases.
Drop any remaining rows with missing CFR

```python
africa_df['new_cases']=africa_df['new_cases'].fillna(0)
africa_df['new_deaths']=africa_df['new_deaths'].fillna(0)
africa_df['total_cases']=africa_df['total_cases'].fillna(0)
africa_df['new_cases_per_million']=africa_df['new_cases_per_million'].fillna(0)
africa_df['new_deaths_per_million']=africa_df['new_deaths_per_million'].fillna(0)
africa_df['cfr']=africa_df['cfr'].fillna((africa_df['total_deaths']/africa_df['total_cases'])*100)
africa_df=africa_df.dropna(subset=['cfr'])
print("Final dataset shape:",africa_df.shape)
print("\nRemaining NaN values:")
print(africa_df.isnull().sum())
```

##11. Descriptive Statistics
Shows basic statistics(mean, standard deviation, min, quartile, max) for key columns in the dataset.

```python
print("DESCRIPTIVE STATISTICS")
print(africa_df[['new_cases','new_deaths','total_cases','total_deaths','new_cases_per_million','new_deaths_per_million','cfr']].describe())
```

##12. Limit CFR Values

Clip CFR values at 100% to avoid unrealistic numbers and show updated statistics.

```python
africa_df['cfr']=africa_df['cfr'].clip(upper=100)
print(africa_df['cfr'].describe())
```

##13. CFR by Country

Calculate the average CFR for each African country,  
then show the top 5 highest, top 5 lowest, and overall mean and median CFR.




```python
cfr_by_country=africa_df.groupby('country')['cfr'].mean().sort_values(ascending=False)
print("AVERAGE CFR BY COUNTRY")
print(cfr_by_country)
print("\n TOP 5 HIGHEST CFR COUNTRIES")
print(cfr_by_country.head(5))
print("\n TOP 5 LOWEST CFR COUNTRIES")
print(cfr_by_country.tail(5))
print("\n OVERALL AFRICA AVERAGE CFR")
print(f"Mean CFR: {cfr_by_country.mean():.2f}%")
print(f"Median CFR: {cfr_by_country.median():.2f}%")
```

##14. Visualize CFR by Country

Create a horizontal bar chart showing the average CFR for each African country.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
cfr_by_country.plot(kind='barh',color='red',alpha=0.7)
plt.title('Average Case Fatality Rate (CFR) by African Country')
plt.xlabel('CFR(%)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

##15. Classify Healthcare Effectiveness

Classify African countries based on their average CFR:  
- Excellent: CFR < 1%  
- Good: CFR 1–2%  
- Moderate: CFR 2–3%  
- Poor: CFR ≥ 3%

```python
cfr_by_country=africa_df.groupby('country')['cfr'].mean().sort_values(ascending=False)
print("HEALTHCARE EFFECTIVENESS CLASSIFICATION")
for country, cfr in cfr_by_country.items():
  if cfr<1:
    effectiveness='Excellent'
  elif cfr<2:
    effectiveness='Good'
  elif cfr<3:
    effectiveness='Moderate'
  else:
    effectiveness='Poor'
    print(f"{country}:CFR={cfr:.2f}%-{effectiveness}")
```

##16. CFR Distribution Summary

Show overall CFR statistics (mean, median, min, max)  
and count countries in each healthcare effectiveness class.

```python
print("CFR DISTRIBUTION SUMMARY")
print(f"Mean CFR:{africa_df['cfr'].mean():.2f}%")
print(f"Median CFR:{africa_df['cfr'].median():.2f}%")
print(f'Min CFR:{africa_df['cfr'].min():.2f}%')
print(f'Max CFR:{africa_df['cfr'].max():.2f}%')

print("\n COUNT BY EFFECTIVENESS CLASS")
excellent_mask=cfr_by_country<1
good_mask=(cfr_by_country>=1) & (cfr_by_country<2)
moderate_mask=(cfr_by_country>=2) & (cfr_by_country<3)
poor_mask=cfr_by_country>=3

excellent_countries=cfr_by_country[excellent_mask]
good_countries=cfr_by_country[good_mask]
moderate_countries=cfr_by_country[moderate_mask]
poor_countries=cfr_by_country[poor_mask]

print(f"Number of countries:{len(moderate_countries)}")
print(moderate_countries)

print(f"Number of countries:{len(poor_countries)}")
print(poor_countries)

total=len(cfr_by_country)
print(f"Total countries analysed:{total}")
print(f"Excellent:{len(excellent_countries)}countries({round(len(excellent_countries)/total*100,1)}%)")
print(f"Good:{len(good_countries)}countries({round(len(good_countries)/total*100,1)}%)")
print(f"Moderate:{len(moderate_countries)}countries({round(len(moderate_countries)/total*100,1)}%)")
print(f"Poor:{len(poor_countries)}countries({round(len(poor_countries)/total*100,1)}%)")

```

##17. Visualize Healthcare Effectiveness

Create a pie chart showing the percentage of countries in each CFR-based effectiveness class (Excellent, Good, Moderate, Poor).

```python
labels=[]
sizes=[]
colors_list=[]
if len(excellent_countries)>0:
  labels.append('Excellent(<1%)')
  sizes.append(len(excellent_countries))
  colors_list.append('green')
  if len(good_countries)>0:
    labels.append('Good(1-2%)')
    sizes.append(len(good_countries))
    colors_list.append('blue')
    if len(moderate_countries)>0:
      labels.append('Moderate(2-3%)')
      sizes.append(len(moderate_countries))
      colors_list.append('orange')
      if len(poor_countries)>0:
        labels.append('Poor(>3%)')
        sizes.append(len(poor_countries))
        colors_list.append('red')
        plt.figure(figsize=(8,8))
        plt.pie(sizes,labels=labels,colors=colors_list,autopct='%1.1f%%',startangle=90)
        plt.title("COVID-19 Healthcare Effectiveness Classification\nAcross Based on Average CFR")
        plt.tight_layout()
        plt.show()
```

##18. Top Countries Analysis

Show the top 10 countries by total cases and total deaths,  
and the top 5 countries with the highest and lowest CFR.

```python
top_cases=africa_df.groupby('country')['total_cases'].max().sort_values(ascending=False).head(10)
print("TOP 10 MOST AFFECTED COUNTRIES")
print(top_cases)

top_deaths=africa_df.groupby('country')['total_deaths'].max().sort_values(ascending=False).head(10)
print("\nTOP 10 COUNTRIES BY TOTAL DEATHS")
print(top_deaths)

print("\n TOP 5 HIGHEST CFR COUNTRIES")
print(poor_countries.head(5))

print("\n TOP 5 LOWEST CFR COUNTRIES")
print(excellent_countries.tail(5) if len(excellent_countries)>0 else cfr_by_country.tail(5))
```

##19. Visualize Top Countries

Create side-by-side bar charts showing the top 10 African countries by total COVID-19 cases and total deaths.

```python
fig,axes=plt.subplots(1,2,figsize=(16,6))
top_cases.plot(kind='bar',ax=axes[0],color='steelblue',alpha=0.8)
axes[0].set_title('Top 10 African Countries\nby Total COVID-19 Cases')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('Total Cases')
axes[0].tick_params(axis='x',rotation=45)

top_deaths.plot(kind='bar',ax=axes[1],color='red',alpha=0.8)
axes[1].set_title('Top 10 African Countries\nby Total COVID-19 Deaths')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Total Death')
axes[1].tick_params(axis='x',rotation=45)

plt.tight_layout()
plt.show()
```

##20. Correlation Analysis

Check the correlation between new COVID-19 cases and new deaths in Africa and interpret the strength of the relationship.

```python
correlation=africa_df['new_cases'].corr(africa_df['new_deaths'])
print("CORRELATION ANALYSIS")
print(f"Pearson Correlation Coefficient:{correlation:.4f}")

if correlation>=0.8:
  interpretation="Very Strong Positive Correlation"
elif correlation>=0.6:
  interpretation="Strong Positive Correlation"
elif correlation>=0.4:
      interpretation="Moderate Positive Correlation"
elif correlation>=0.2:
     interpretation="Weak Positive Correlation"
else:
    interpretation="Very Weak or No Correlation"
    print(f"Interpretation:{interpretation}")
```

##21. Visualize Correlation

Scatter plot showing the relationship between daily new cases and new deaths across Africa, with the Pearson correlation coefficient annotated.

```python
plt.figure(figsize=(12,8))
plt.scatter(africa_df['new_cases'],africa_df['new_deaths'],alpha=0.3,color='purple',s=5)
plt.title('Correlation Between Daily New Cases \nand New Deaths Across Africa')
plt.xlabel('New Cases')
plt.ylabel('New Deaths')
plt.annotate(f'r={correlation:.4f}',xy=(0.7,0.9),xycoords='axes fraction',fontsize=12,color='black',bbox=dict(boxstyle='round',facecolor='wheat',alpha=0.5))
plt.tight_layout
plt.show()
```

##22. Time Series Summary

Convert the date column to datetime and summarize daily new COVID-19 cases across Africa, including start, end, peak, and average values.

```python
africa_df['date']=pd.to_datetime(africa_df['date'])
time_series=africa_df.groupby('date')['new_cases'].sum()
print("\n TIME SERIES SUMMARY")
print(f"Start Date:{time_series.index.min()}")
print(f"End Date:{time_series.index.max()}")
print(f"Peak date:{time_series.idxmax()}")
print(f"Peak new cases:{time_series.max():,.0f}")
print(f"Average daily new cases:{time_series.mean():,.0f}")
```

##23. Visualize Daily Cases Over Time

Line chart showing daily new COVID-19 cases across Africa, with the peak day highlighted.

```python
plt.figure(figsize=(14,6))
time_series.plot(color='green',linewidth=1)
plt.title('Daily New COVID-19 Cases Across Africa Over Time')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.axvline(x=time_series.idxmax(),color='red',linestyle='--',label=f'Peak:{time_series.idxmax().date()}')
plt.legend()
plt.tight_layout()
plt.show()
```
